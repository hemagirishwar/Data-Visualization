##  bar graph

```python
import matplotlib.pyplot as plt

x=['python','c++','c']
y=[85,70,60]
plt.xlabel('language',fontsize=20)
plt.ylabel('number',fontsize=15)
plt.title('programing',fontsize=20)
a=['red','green','blue']
#width:space between each bar
#align(edge,center):aligns the name of the bar in sepcfied align command
#edgecolor:defines the color of the outline of the bar
#linewidth:size of the edgecolor
#linestyle:style of the edgecolor
#alpha:visiblityof the bar
plt.bar(x,y,width=0.4,color=a,align='edge',edgecolor='yellow',linewidth=3,linestyle=':',alpha=0.5)
plt.legend('a')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=['python','c++','c']
y=[85,70,60]
z=[30,60,20]
width=0.2
a=np.arange(len(x))
b=[i+width for i in a]
plt.xlabel('language',fontsize=20)
plt.ylabel('number',fontsize=15)
plt.title('programing',fontsize=20)
#lable:for describing the bar graph
plt.bar(a,y,width,color='orange',label='1')
plt.bar(b,z,width,color='blue',label='2')
#xtrick:used to give the names for the bar graph and width/2 aligns the name  at the center of the two bar graph
plt.xticks(a+width/2,x,rotation=20)
plt.legend()
plt.show()
```

## scatter plot

```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5,6,7]
y=[2,5,6,8,9,6,3]
a=[20,35,62,48,90,39,50]
size=[300,200,400,300,250,600,500]
plt.title('scatter plot',fontsize=20)
plt.xlabel('day',fontsize=20)
plt.ylabel('number',fontsize=20)
#s:for specfying the size of the dot
#marker:giving the shape to the points in graph
#c:is used for defining the color
#cmap:is a collection of color in which coloe codes are given
plt.scatter(x,y,c=a,cmap='viridis',s=size,marker='*')
t=plt.colorbar()
t.set_label("color bar",fontsize=15)
plt.show()
```

## histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random


np.random.seed(50)
a=np.random.randint(1,50,size=(50))
b=[0,10,20,30,40,50]
plt.title('histogram',fontsize=20)
plt.xlabel('python',fontsize=20)
plt.ylabel('numbers',fontsize=20)
#bins:used to align the graph in the specified values,the values should be in assending orde
#cumulative=-1:gives the graph in descending order
#cumulative=1:gives the graph in ascending order
#bottom:starting point of the values in the graph depending on the orientation
#histtype:defines the type of the hystogram
#orientation:vertical or horizontal graph
plt.hist(a,color='r',edgecolor='black',linewidth=5,bins=b,cumulative=-1,bottom=10,histtype='step',orientation='horizontal')
plt.show()
 
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random


np.random.seed(50)
a=np.random.randint(1,50,size=(50))
b=[0,10,20,30,40,50]
plt.title('histogram',fontsize=20)
plt.xlabel('python',fontsize=20)
plt.ylabel('numbers',fontsize=20)
#range:specfies the start and end points of the axis depending on the orientation
plt.hist(a,'auto',edgecolor='black',linewidth=5,range=(0,200),orientation='vertical')
plt.show()
 
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random


np.random.seed(50)
a=np.random.randint(1,50,size=(50))
b=[0,10,20,30,40,50]
plt.title('histogram',fontsize=20)
plt.xlabel('python',fontsize=20)
plt.ylabel('numbers',fontsize=20)
#rwidth:defines the width between each bar
plt.hist(a,color='r',edgecolor='black',linewidth=2,bins=b,rwidth=0.8,label='python')
#axvline:draws a line in the specfied value
plt.axvline(30,color='blue',label='line')
plt.legend()
#grid:defines the grid
plt.grid()
plt.show()
```

## pie plot

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.1,0.2,0.0,0.0]
c=['r','b','g','y']
plt.title('pie plot',fontsize=20,color='brown')
#explode=used to seperate individual pie areas in specfied values
#autocpt=used to give the values of the bar graph ,i and f can be used i represents integer and f represents float values,any valus after 0. defines the number of values after the point
#the syntax for autopct is "%0.(values)(i or f)%%"
#shadow=gives shadow to the pie graph,its value can be True and False
#radius=defines the radius to the pie graph
#labeldistance=defines the radius for the labels
#startangle=defines the angle where the pie plot should start
#textprops=is uesd to pass fontsize,it should be mentioned in dictionary
#counterclock=used to rotate the names of the individual pie area in clock or anti-clock wise,its values should be True or False
#wedgeprops=is used to define the properties of the bar graph,it should be given in a dictionary
#edgecolor=used to give the color to the edges
#rotatelabel=used to ratate th labels,its values must be True or False
plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',shadow=True,radius=0.8,labeldistance=1.1,startangle=180,textprops={'fontsize':20},counterclock=False,wedgeprops={'linewidth':1,'edgecolor':'lightgreen'},rotatelabels=True)
#loc=gives the location to the lables
plt.legend(loc=2)
plt.show()
```

## multiple pie plot

##  stem plot 

```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5]
y=[5,6,7,8,9]
#linefmt=used to specify the type of line
#markerfmt=used to give the shape to the point
#bottom=shifts the bottom line in the specfied value
#basefmt=used to specify the color of the line
plt.stem(x,y,linefmt=':',markerfmt='r*',bottom=10,basefmt='green',label='python',orientation='horizontal')
plt.legend()
plt.show()
```

## box plot

```python
import matplotlib.pyplot as plt

x=[50,40,80,60,30,70,50,120]
#notch=gives the shape of notch to the boxplot,its values must be given as true or false
#vert=aligns the graph in vertical or horizontal way,its values must be given as true or false
#patch_artist=give color inside the box,its values must be given as true or false
#showmeans=shoes the position of mean in the graph,its values must be given as true or false
#sym=gives color and shape to the outler
#boxprops=used to define the color of the box
#capprops=used to define the color of the horizontal  min and max lines.the horizontal lines extending above and below are called max and min caps
#whiskerprops=used to define the color of the vertical min amd max lines.the vertical lines extending above and below are called min max hiskers
plt.boxplot(x,notch=False,vert=True,widths=0.1,labels=['python'],patch_artist=True,showmeans=True,sym='g+',boxprops={'color':'red'},capprops={'color':"green"},whiskerprops={'color':"brown"})
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[50,40,80,60,30,70,50,120]
#whis=extends the above line the the boxplot to join the outler(circle outside the graph)
plt.boxplot(x,whis=10)
plt.legend()
plt.show()
```

##  stack(or)area plot

```python
import matplotlib.pyplot as plt

x=[5,4,6,3,2,9]
x.sort()
y1=[5,6,3,2,4,1]
y1.sort()
y2=[5,6,2,3,6,10]
y2.sort()
y3=[2,3,6,5,9,4]
y3.sort()
l=['python','c++','c']
plt.stackplot(x,y1,y2,y3,labels=l,colors=['r','g','y'])
plt.legend(loc=2)
plt.show()
```

##  text

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,5,4,2,1]
plt.plot(x,y)
plt.title('vs-code',fontsize=15)
plt.xlabel('days',fontsize=10)
plt.ylabel('python',fontsize=10)
#bbox=attributes of the box
plt.text(2.2,5,'hello',fontsize=10,style="italic",bbox={'facecolor':'y',"edgecolor":"r"})
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,5,4,2,1]
plt.plot(x,y)
plt.title('vs-code',fontsize=15)
plt.xlabel('days',fontsize=10)
plt.ylabel('python',fontsize=10)
#arrowprops=arrow properties
#xy=axis of the arrow
#xytext=axis of the text
plt.annotate('hellow',xy=(2.2,5),xytext=(4,4),arrowprops={'facecolor':'r',"shrink":100})
plt.show()
```
