##  line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var=[1,2,3,4,5,6,7]
var_1=[2,3,5,6,2,4,5]

df=pd.DataFrame({'var':var,'var_1':var_1})

sns.lineplot(x='var',y='var_1',data=df)
plt.title("hello")
plt.show()
```

##  histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=pd.read_csv(r'C:\Users\hemag\OneDrive\Desktop\my.py\my.py\penguins.csv')

#kde:kernal density,a curved line
#rug:shows lines on the x-axis below the kde
sns.displot(df['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True,color='r')
plt.show()
```

##  bar plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=pd.read_csv(r'C:\Users\hemag\OneDrive\Desktop\my.py\my.py\penguins.csv')


order_1=['Dream','Torgersen','Biscoe']
#order:sequence
#ci:height of the line above the bar graph,its range is from 0 to 100
#errcolor:defines the color of the line above the graph
#errwidth:defines the width of the line above the graph
sns.set(style='darkgrid')
sns.barplot(x=df['island'],y=df['bill_length_mm'],hue=df['sex'],order=order_1,ci=100,orient='v',errcolor='g',errwidth=5)
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=pd.read_csv(r'C:\Users\hemag\OneDrive\Desktop\my.py\my.py\penguins.csv')

#saturation:intensity of the bars
sns.barplot(x=df['bill_depth_mm'],y=df['bill_length_mm'],orient='v',palette='icefire',saturation=0.3)
plt.show()
```

##  strip plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=sns.load_dataset("tips")
#jitter:width between each points
sns.stripplot(x='day',y='total_bill',hue='sex',data=df,palette="rocket_r",linewidth=1,edgecolor="black",jitter=0.2,size=10)
plt.legend(loc=2)
plt.show()
```

##  box plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=sns.load_dataset("tips")

sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="sex",order=["Fri","Sun","Thur","Sat"],showmeans=True,meanprops={"marker":"+","markeredgecolor":"r"},linewidth=5,orient="v")
plt.show()
```
